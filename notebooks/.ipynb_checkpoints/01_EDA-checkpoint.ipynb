## 0. Imports & Setup

```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import warnings

warnings.filterwarnings("ignore")
pd.set_option("display.max_columns", None)
pd.set_option("display.float_format", "{:.4f}".format)

print("✅ All libraries loaded successfully.")
```

## 1. Load Dataset

The raw CSV uses `;` as a separator and stores `?` for missing values.  
We load with these parameters and immediately take a first look.


```python
DATA_PATH = "../data/household_power_consumption.csv"

df_raw = pd.read_csv(
    DATA_PATH,
    sep=";",
    na_values=["?"],         # '?' represents missing in this dataset
    low_memory=False
)

print(f"Shape  : {df_raw.shape[0]:,} rows × {df_raw.shape[1]} columns")
print(f"Columns: {list(df_raw.columns)}")
df_raw.head()
```

## 2. Dataset Overview

```python
df_raw.info()
```

```python
# Quick summary statistics on the raw (object-typed) load
df_raw.describe(include="all").T
```

## 3. Data Cleaning

### 3.1 Combine Date + Time → DatetimeIndex
All columns except `Sub_metering_3` were loaded as `object`; we convert them to
numeric after building a proper datetime index.


```python
# Combine Date and Time strings into a single Datetime column
df = df_raw.copy()
df["Datetime"] = pd.to_datetime(
    df["Date"] + " " + df["Time"],
    dayfirst=True            # date format is dd/mm/yyyy
)

df = df.sort_values("Datetime").set_index("Datetime")
df.drop(columns=["Date", "Time"], inplace=True)

# Convert all power/voltage columns to float
numeric_cols = [
    "Global_active_power",
    "Global_reactive_power",
    "Voltage",
    "Global_intensity",
    "Sub_metering_1",
    "Sub_metering_2",
    "Sub_metering_3",
]

for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors="coerce")

print("✅ Datetime index set | Numeric columns converted")
print(f"Date range : {df.index.min()}  →  {df.index.max()}")
df.head()
```

### 3.2 Missing Value Analysis

```python
missing = df.isnull().sum()
missing_pct = (missing / len(df) * 100).round(3)

missing_df = pd.DataFrame({
    "Missing Count": missing,
    "Missing %"    : missing_pct
}).sort_values("Missing %", ascending=False)

print(missing_df)
print(f"\nTotal missing cells : {missing.sum():,} / {df.size:,}  "
      f"({missing.sum()/df.size*100:.2f}%)")
```

```python
# Visualise missingness pattern
fig = px.bar(
    missing_df[missing_df["Missing Count"] > 0].reset_index(),
    x="index", y="Missing %",
    text="Missing %",
    title="Missing Values per Column (%)",
    labels={"index": "Column"},
    color="Missing %",
    color_continuous_scale="Reds",
)
fig.update_traces(texttemplate="%{text:.2f}%", textposition="outside")
fig.update_layout(title_x=0.5, height=450, coloraxis_showscale=False)
fig.show()
```

### 3.3 Handle Missing Values

Missing values are approximately **1.25 %** of the dataset — they tend to appear
in contiguous blocks (sensor outages). We use **forward-fill** to propagate the
last known reading, which is the standard approach for sensor time series.


```python
df_clean = df.copy()
df_clean.ffill(inplace=True)          # forward-fill sensor gaps
df_clean.bfill(inplace=True)          # back-fill any leading NaNs

remaining_missing = df_clean.isnull().sum().sum()
print(f"✅ Missing values after fill: {remaining_missing}")
print(f"Final clean dataset shape  : {df_clean.shape}")
```

## 4. Feature Engineering — Time Attributes

We extract temporal features that will be crucial for both EDA and later modelling.


```python
df_clean["Year"]        = df_clean.index.year
df_clean["Month"]       = df_clean.index.month
df_clean["Month_Name"]  = df_clean.index.month_name()
df_clean["DayOfWeek"]   = df_clean.index.dayofweek          # 0=Mon … 6=Sun
df_clean["Day_Name"]    = df_clean.index.day_name()
df_clean["Hour"]        = df_clean.index.hour
df_clean["IsWeekend"]   = (df_clean.index.dayofweek >= 5).astype(int)
df_clean["Quarter"]     = df_clean.index.quarter
df_clean["Season"]      = df_clean["Month"].map({
    12:"Winter", 1:"Winter", 2:"Winter",
    3:"Spring",  4:"Spring",  5:"Spring",
    6:"Summer",  7:"Summer",  8:"Summer",
    9:"Autumn", 10:"Autumn", 11:"Autumn"
})

print("✅ Temporal features added")
df_clean[["Year","Month","DayOfWeek","Hour","IsWeekend","Season"]].head()
```

## 5. Target Variable Analysis — `Global_active_power`

The target variable is the **total active power** consumption of the household,
measured in kilowatts (kW) at each minute.


```python
gap = df_clean["Global_active_power"]

print("=== Global Active Power — Summary Statistics ===")
print(f"  Mean   : {gap.mean():.4f} kW")
print(f"  Median : {gap.median():.4f} kW")
print(f"  Std Dev: {gap.std():.4f} kW")
print(f"  Min    : {gap.min():.4f} kW")
print(f"  Max    : {gap.max():.4f} kW")
print(f"  Skew   : {gap.skew():.4f}")
```

```python
fig = px.histogram(
    df_clean,
    x="Global_active_power",
    nbins=100,
    title="Distribution of Global Active Power (kW)",
    labels={"Global_active_power": "Active Power (kW)"},
    color_discrete_sequence=["#2196F3"],
    marginal="box",
)
fig.update_layout(title_x=0.5, height=500)
fig.show()
```

**Interpretation:**  
The distribution is **right-skewed** — most minutes have low consumption
(0–2 kW, e.g., standby loads), with a long tail of high-consumption spikes
from appliances like water heaters, ovens, and air conditioners.


## 6. Time-Series Visualisation

### 6.1 Full Series (Daily Average)

```python
daily_avg = df_clean["Global_active_power"].resample("D").mean().reset_index()
daily_avg.columns = ["Date", "Avg_Power_kW"]

fig = px.line(
    daily_avg,
    x="Date", y="Avg_Power_kW",
    title="Daily Average Global Active Power (Dec 2006 – Nov 2010)",
    labels={"Avg_Power_kW": "Avg Power (kW)", "Date": ""},
    color_discrete_sequence=["#1565C0"],
)
fig.update_layout(title_x=0.5, height=450)
fig.show()
```

**Interpretation:**  
- Clear **seasonal pattern** — consumption peaks in winter and dips in summer, consistent with heating demand in France.  
- A few brief gaps (flat lines) correspond to the ~1.25 % missing-value periods we filled.


### 6.2 Seasonal Pattern — Monthly Average

```python
month_order = ["January","February","March","April","May","June",
               "July","August","September","October","November","December"]

monthly = (
    df_clean.groupby(["Year","Month_Name"])["Global_active_power"]
    .mean()
    .reset_index()
)
monthly["Month_Name"] = pd.Categorical(monthly["Month_Name"],
                                        categories=month_order, ordered=True)
monthly = monthly.sort_values(["Year","Month_Name"])

fig = px.line(
    monthly, x="Month_Name", y="Global_active_power",
    color="Year",
    title="Monthly Average Active Power by Year",
    labels={"Global_active_power": "Avg Power (kW)", "Month_Name": "Month"},
    markers=True,
    color_discrete_sequence=px.colors.qualitative.Set2,
)
fig.update_layout(title_x=0.5, height=480)
fig.show()
```

### 6.3 Daily Pattern — Hour of Day

```python
hourly = (
    df_clean.groupby(["Hour", "IsWeekend"])["Global_active_power"]
    .mean()
    .reset_index()
)
hourly["Day_Type"] = hourly["IsWeekend"].map({0: "Weekday", 1: "Weekend"})

fig = px.line(
    hourly, x="Hour", y="Global_active_power",
    color="Day_Type",
    title="Average Power Consumption by Hour of Day (Weekday vs Weekend)",
    labels={"Global_active_power": "Avg Power (kW)", "Hour": "Hour of Day"},
    markers=True,
    color_discrete_sequence=["#1976D2", "#E53935"],
)
fig.update_layout(title_x=0.5, height=450,
                  xaxis=dict(tickmode="linear", dtick=2))
fig.show()
```

**Interpretation:**  
- **Morning ramp** (~6–9 AM): breakfast and morning routines.  
- **Evening peak** (~6–9 PM): dinner, TV, and appliance use after work.  
- **Weekend pattern** is flatter and starts later, consistent with sleeping in.


### 6.4 Weekly Pattern — Day of Week

```python
day_order = ["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"]

weekly = (
    df_clean.groupby("Day_Name")["Global_active_power"]
    .agg(["mean", "std"])
    .reindex(day_order)
    .reset_index()
)

fig = go.Figure([
    go.Bar(
        x=weekly["Day_Name"],
        y=weekly["mean"],
        error_y=dict(type="data", array=weekly["std"], visible=True, color="#999"),
        marker_color=["#42A5F5"]*5 + ["#EF5350","#EF5350"],
        name="Avg Power (kW)",
    )
])
fig.update_layout(
    title="Average Active Power by Day of Week (±1 std)",
    xaxis_title="Day of Week", yaxis_title="Avg Power (kW)",
    title_x=0.5, height=450, showlegend=False
)
fig.show()
```

### 6.5 Seasonal Boxplot

```python
season_order = ["Winter","Spring","Summer","Autumn"]
df_season = df_clean[df_clean["Season"].isin(season_order)].copy()

# Sample for plotting speed
df_sample = df_season.sample(n=50_000, random_state=42)

fig = px.box(
    df_sample, x="Season", y="Global_active_power",
    category_orders={"Season": season_order},
    color="Season",
    title="Active Power Distribution by Season",
    labels={"Global_active_power": "Active Power (kW)"},
    color_discrete_map={
        "Winter":"#1E88E5","Spring":"#43A047",
        "Summer":"#FB8C00","Autumn":"#8E24AA"
    },
)
fig.update_layout(title_x=0.5, height=480, showlegend=False)
fig.show()
```

## 7. Sub-Meter Contribution Analysis

Three sub-meters track specific appliance groups:

| Sub-meter | Appliances |
|-----------|-----------|
| Sub_metering_1 | Kitchen (microwave, dishwasher, oven) |
| Sub_metering_2 | Laundry (washing machine, dryer, fridge) |
| Sub_metering_3 | Water heater & air conditioner |


```python
# Convert sub-metering from Wh to kWh for comparison with power readings
# Global_active_power is in kW (per minute = kWh/60)
df_clean["Sub1_kWh"] = df_clean["Sub_metering_1"] / 1000
df_clean["Sub2_kWh"] = df_clean["Sub_metering_2"] / 1000
df_clean["Sub3_kWh"] = df_clean["Sub_metering_3"] / 1000

# Total metered sub-consumption vs. total (in energy-per-minute terms)
df_clean["Total_Sub_kWh"] = df_clean[["Sub1_kWh","Sub2_kWh","Sub3_kWh"]].sum(axis=1)
df_clean["Unmetered_kW"]  = (
    df_clean["Global_active_power"] - df_clean["Total_Sub_kWh"]
).clip(lower=0)

total_energy = df_clean["Global_active_power"].sum()
sub1_share   = df_clean["Sub1_kWh"].sum() / total_energy * 100
sub2_share   = df_clean["Sub2_kWh"].sum() / total_energy * 100
sub3_share   = df_clean["Sub3_kWh"].sum() / total_energy * 100
unmeter_share= df_clean["Unmetered_kW"].sum() / total_energy * 100

print("=== Energy Share per Sub-meter ===")
print(f"  Kitchen (Sub1)   : {sub1_share:.1f}%")
print(f"  Laundry (Sub2)   : {sub2_share:.1f}%")
print(f"  HVAC    (Sub3)   : {sub3_share:.1f}%")
print(f"  Unmetered        : {unmeter_share:.1f}%")
```

```python
labels = ["Kitchen (Sub1)","Laundry (Sub2)","HVAC (Sub3)","Unmetered/Other"]
values = [sub1_share, sub2_share, sub3_share, unmeter_share]

fig = px.pie(
    names=labels, values=values,
    title="Energy Share by Sub-meter Category",
    color_discrete_sequence=px.colors.qualitative.Set2,
    hole=0.4
)
fig.update_layout(title_x=0.5, height=480)
fig.show()
```

```python
# Monthly sub-meter averages
monthly_sub = (
    df_clean.resample("M")[["Sub1_kWh","Sub2_kWh","Sub3_kWh"]]
    .mean()
    .reset_index()
    .melt(id_vars="Datetime", var_name="Sub_Meter", value_name="Avg_kWh")
)
monthly_sub["Sub_Meter"] = monthly_sub["Sub_Meter"].map({
    "Sub1_kWh":"Kitchen","Sub2_kWh":"Laundry","Sub3_kWh":"HVAC"
})

fig = px.line(
    monthly_sub, x="Datetime", y="Avg_kWh",
    color="Sub_Meter",
    title="Monthly Average Sub-meter Consumption Over Time",
    labels={"Avg_kWh": "Avg Consumption (kWh/min)", "Datetime": ""},
    markers=False,
    color_discrete_sequence=["#E53935","#43A047","#1E88E5"],
)
fig.update_layout(title_x=0.5, height=450)
fig.show()
```

## 8. Correlation Analysis

```python
corr_cols = [
    "Global_active_power","Global_reactive_power","Voltage",
    "Global_intensity","Sub_metering_1","Sub_metering_2","Sub_metering_3"
]
corr = df_clean[corr_cols].corr()

fig = px.imshow(
    corr,
    text_auto=".2f",
    color_continuous_scale="RdBu_r",
    zmin=-1, zmax=1,
    title="Pearson Correlation Matrix",
)
fig.update_layout(title_x=0.5, height=600)
fig.show()
```

**Key Correlations:**
- `Global_active_power` ↔ `Global_intensity`: **~0.99** — as expected from Ohm's law; intensity drives power.
- `Global_active_power` ↔ `Sub_metering_3`: moderate positive — HVAC is a major load driver.
- `Voltage` ↔ `Global_intensity`: negative — higher current draw slightly depresses mains voltage.


## 9. Outlier Detection

```python
from scipy import stats

# Z-score outliers on the target variable
z_scores = np.abs(stats.zscore(df_clean["Global_active_power"].dropna()))
outlier_count = (z_scores > 3).sum()
outlier_pct   = outlier_count / len(df_clean) * 100

print(f"Outliers (|z| > 3): {outlier_count:,}  ({outlier_pct:.2f}% of dataset)")

# IQR method
Q1 = df_clean["Global_active_power"].quantile(0.25)
Q3 = df_clean["Global_active_power"].quantile(0.75)
IQR = Q3 - Q1
iqr_outliers = ((df_clean["Global_active_power"] < Q1 - 1.5*IQR) |
                (df_clean["Global_active_power"] > Q3 + 1.5*IQR)).sum()
print(f"Outliers (IQR method): {iqr_outliers:,}  ({iqr_outliers/len(df_clean)*100:.2f}%)")
```

```python
# Box plot for all numeric features
fig = px.box(
    df_clean[corr_cols].sample(30_000, random_state=42).melt(
        var_name="Feature", value_name="Value"),
    x="Feature", y="Value",
    color="Feature",
    title="Boxplot of All Numeric Features (30k sample)",
    color_discrete_sequence=px.colors.qualitative.Pastel,
)
fig.update_layout(title_x=0.5, height=500, showlegend=False)
fig.show()
```

## 10. Multi-Resolution Aggregates

```python
# We create hourly and daily resampled datasets — used as input for the ML notebook
df_hourly = df_clean[numeric_cols].resample("H").mean()
df_daily  = df_clean[numeric_cols].resample("D").mean()

print(f"Hourly dataset shape : {df_hourly.shape}")
print(f"Daily  dataset shape : {df_daily.shape}")

# Save for use by the modelling notebook
df_hourly.to_csv("../data/hourly_clean.csv")
df_daily.to_csv("../data/daily_clean.csv")
df_clean.to_parquet("../data/minute_clean.parquet")

print("\n✅ Saved hourly_clean.csv | daily_clean.csv | minute_clean.parquet")
```
